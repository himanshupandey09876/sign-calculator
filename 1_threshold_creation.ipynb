```python
# Define input and output folders
input_folder = r'C:\Users\himan\Downloads\cs771ml\sign-calculator-main\blackoutimages\N'
output_folder = r'C:\Users\himan\Downloads\cs771ml\sign-calculator-main\blackoutimages\N6'
```

```python
import cv2
import os

def threshold_image(image_path, output_path):
    """Apply binary threshold to an image and save the result."""
    # Read the image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Apply binary threshold
    _, binary = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
    
    # Save the thresholded image
    cv2.imwrite(output_path, binary)

def process_images(input_folder, output_folder):
    """Process all images in the input folder and save them to the output folder."""
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Process each image in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".png") or filename.endswith(".jpg"):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            threshold_image(input_path, output_path)
            print(f"Processed {filename}")



# Process the images
process_images(input_folder, output_folder)
```
